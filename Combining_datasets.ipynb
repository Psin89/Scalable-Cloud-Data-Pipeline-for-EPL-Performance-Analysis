```python
import pandas as pd

# --- 1. Load datasets ---
df1 = pd.read_csv(r"C:\Users\Aditya Singh\Desktop\Study\CloudTech\Project\data\E0.csv")
df2 = pd.read_csv(r"C:\Users\Aditya Singh\Desktop\Study\CloudTech\Project\data\E0 (1).csv")
df3 = pd.read_csv(r"C:\Users\Aditya Singh\Desktop\Study\CloudTech\Project\data\E0 (2).csv")
df4 = pd.read_csv(r"C:\Users\Aditya Singh\Desktop\Study\CloudTech\Project\data\E0 (3).csv")

datasets = [df1, df2, df3, df4]

for i, df in enumerate(datasets):
    df['match_datetime'] = pd.to_datetime(
        df['Date'] + " " + df['Time'],
        dayfirst=True,
        format="%d/%m/%Y %H:%M"
    )
    datasets[i] = df

    # Save back
    datasets[i] = df

# --- 3. Merge all datasets on match_datetime + teams ---
# Adjust column names if necessary (e.g., HomeTeam, AwayTeam)
merged = datasets[0]
for df in datasets[1:]:
    merged = merged.merge(
        df,
        on=['match_datetime', 'HomeTeam', 'AwayTeam'],
        how='outer',
        suffixes=('', '_dup')  # prevents name clashes
    )

# --- 4. Drop Date + Time columns ---
cols_to_drop = [col for col in ['Date', 'Time'] if col in merged.columns]
merged.drop(columns=cols_to_drop, inplace=True)

# --- 5. Export ---
merged.to_csv(r"C:\Users\Aditya Singh\Desktop\Study\CloudTech\Project\data\SinceCovidPL.csv", index=False)
print("Merge complete! Saved as SinceCovidPL.csv")
```
